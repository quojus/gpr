```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
tsand_er = 4
tsand_o = 0.1
wsand_er = 25
wsand_o = 1
tton_er = 18
tton_o = 100
wton_er = 30
wton_o = 2000
e0 = 8.854*10**(-12)
#  mü = mü0
mü = 4*np.pi*10**(-7)

tsand_e =  tsand_er*e0
wsand_e = wsand_er*e0
tton_e = tton_er*e0
wton_e = wton_er*e0

fmin =4
fmax = 11

#schrite
ite= 10*3

f = np.logspace(fmin,fmax,ite)
```

```python
####
def v_(o,e):
    v = np.zeros(len(f))
    #alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        v[i] =  w[i]/np.real(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i]))) 
        #alfa[i] = -np.imag(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i])))
    
    return(v)
    
def alfa_(o,e):
    #v = np.zeros(len(f))
    alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        #v[i] =  w[i]/np.real(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i]))) 
        alfa[i] = -np.imag(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i])))
    
    return(alfa)   
####
def v_approx(o,e):
    v = np.zeros(len(f))
    alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        v[i] = 1/np.sqrt(e*mü)
    
    return(v)

def alfa_approx(o,e):
    #v = np.zeros(len(f))
    alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        #v[i] =  w[i]/np.real(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i]))) 
        alfa[i] = (mü*o*1/np.sqrt(e*mü))/2
    
    return(alfa)   

####
def v_quasi(o,e):
    v = np.zeros(len(f))
    alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        v[i] = np.sqrt(2*w[i]/(o*mü))
    
    return(v)

def alfa_quasi(o,e):
    #v = np.zeros(len(f))
    alfa = np.zeros(len(f))
    w = 2*np.pi*f
    
    for i in range(len(f)):
        
        
        #v[i] =  w[i]/np.real(np.sqrt(complex(e*w[i]**2*mü,o*mü*w[i]))) 
        alfa[i] = np.sqrt((o*mü*w[i])/2)
    
    return(alfa)   
        
```

```python
t1_er = 4
t1_o = 0.1
w1_er = 4
w1_o = 1

t1_e =  t1_er*e0
w1_e = w1_er*e0



#
plt.figure(figsize = [25, 10])
plt.subplot(1,2,1)
plt.ylabel('Phasengeschwindigkeit')
plt.xlabel('frequenz')
plt.xscale('log')
plt.yscale('log')
plt.plot(f,v_(tsand_o,tsand_e),label = 'trockren sand')
plt.plot(f,v_(wsand_o,wsand_e),label = 'nasser sand')  
plt.plot(f,v_(tton_o,tton_e),label = 'trocken sand')  
plt.plot(f,v_(wton_o,wton_e),label = 'nasser ton')  
plt.legend()

plt.subplot(1,2,2)
plt.ylabel('Dämpfungskonstante')
plt.xlabel('frequenz')
plt.xscale('log')
plt.yscale('symlog')
plt.plot(f,alfa_(tsand_o,tsand_e),label = 'trockren sand')
plt.plot(f,alfa_(wsand_o,wsand_e),label = 'nasser sand')

plt.plot(f,alfa_(t1_o,t1_e),label = 'trockren 1')
plt.plot(f,alfa_(w1_o,w1_e),label = 'nasser 1')
#plt.plot(f,alfa_(tton_o,tton_e),label = 'trocken ton')
#plt.plot(f,alfa_(wton_o,wton_e),label = 'nasser ton')
plt.legend()

```
